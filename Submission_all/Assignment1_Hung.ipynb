# Calculate add, subtract, multiply, and power element-wise

```python
a = np.full((1, 2), 2)
b = np.full((1, 2), 3)
print(np.add(a, b))
print(np.subtract(a, b))
print(np.multiply(a, b))
print(np.power(a, b))
```

```python
a = tf.fill((1, 2), 2)
b = tf.fill((1, 2), 3)
print(tf.add(a, b))
print(tf.subtract(a, b))
print(tf.multiply(a, b))
print(tf.pow(a, b))
```

```python
a = torch.full((1, 2), 2)
b = torch.full((1, 2), 3)
print(torch.add(a, b))
print(torch.sub(a, b))
print(torch.mul(a, b))
print(torch.pow(a, b))
```

# Get the sum, max, min, and mean of the array elements over a given axis

```python
arr = np.reshape(np.arange(20),(4,-1))
print(np.sum(arr, axis=0))
print(np.max(arr, axis=0))
print(np.min(arr, axis=0))
print(np.mean(arr, axis=0))
```

```python
arr = tf.reshape(tf.range(20, dtype=tf.float32),(4,-1))
print(tf.reduce_sum(arr, axis=0))
print(tf.reduce_max(arr, axis=0))
print(tf.reduce_min(arr, axis=0))
print(tf.reduce_mean(arr, axis=0))
```

```python
arr = torch.reshape(torch.arange(20, dtype=torch.float32),(4,-1))
print(torch.sum(arr, dim=0))
print(torch.max(arr, dim=0))
print(torch.min(arr, dim=0))
print(torch.mean(arr, dim=0))
```
